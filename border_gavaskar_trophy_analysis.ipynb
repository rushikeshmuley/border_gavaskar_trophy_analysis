```python
Runs=[]
Wics=[]
run_wic=df['team total'].str.split("/").tolist()
for runs,wics in run_wic:
  Runs.append(runs)
  Wics.append(wics)
```

```python
df["RUNS"]=Runs
df["WICKETS"]=Wics
df.drop("team total",axis=1)
df["RUNS"]=df["RUNS"].astype('int32')
df["WICKETS"]=df["WICKETS"].astype('int32')
```
